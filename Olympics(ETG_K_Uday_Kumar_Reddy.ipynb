##  2. Which sport is having most number of Medals so far?

```python
df["Medal"].unique()
```

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (10,5))
```

## 3. Which Athlete is having most number of Medals so far?

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (15,5))
```

## 4.Which Event is most popular in terms of number of players??

```python
df.groupby('Discipline').count()['Gender'].sort_values(ascending = False).head().plot.bar(figsize = (15,5))
```

## 5.Which Sport is having most number of Gold Medals so far??(Top 5)

```python
data = []

for Gold in df['Sport'].unique():
  data.append([Gold , len(df[df['Sport']  == Gold])])

pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
```
